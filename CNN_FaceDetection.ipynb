```python
import dlib
import cv2 
from google.colab.patches import cv2_imshow #importação para poder mostrar imagem
```

```python
faceset = cv2.imread('/content/drive/MyDrive/Visão Computacional/Images/FaceDataSet.jpg')
```

```python
cv2_imshow(faceset)
```

Para utilizar o **CNN** troque o ambiente para **GPU**

```python
detect_face_cnn = dlib.cnn_face_detection_model_v1('/content/drive/MyDrive/Visão Computacional/Weights/mmod_human_face_detector.dat')
```

```python
deteccoes = detect_face_cnn(faceset , 1)
for face in deteccoes: # c representando confiança da detecção das faces
  l , t , r, b ,c = face.rect.left(),face.rect.top(),face.rect.right(),face.rect.bottom(), face.confidence
  print(c)
  cv2.rectangle(faceset,(l, t), (r,b),(0, 255, 100), 2)
cv2_imshow(faceset)
```
